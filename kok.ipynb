```python
import os
import glob

# Specify the path to the directory you want to parse
folder_path = '/home/rikitwiki/Desktop/input_images (copy)/'

# Create a pattern to match all .txt files
pattern = os.path.join(folder_path, '*.txt')

# Use glob.glob to find all files in the folder that match the pattern
txt_files = glob.glob(pattern)

# Loop through the list of txt files and remove each one
for file_path in txt_files:
    try:
        os.remove(file_path)
        print(f'Removed: {file_path}')
    except Exception as e:
        print(f'Error removing {file_path}: {e}')
```
